# Run Backend in Colab (ngrok)
Installs deps, launches FastAPI, exposes public URL, and ingests the dataset.

```python
%pip -q install -r /mnt/data/ikarus-ai-product-app/backend/requirements.txt
print('✅ Deps installed')
```

```python
from pyngrok import ngrok
public_url = ngrok.connect(8000, bind_tls=True).public_url
print('⚡ Public URL:', public_url)
```

```python
import subprocess, sys
proc = subprocess.Popen([sys.executable, '-m', 'uvicorn', 'app.main:app', '--host','0.0.0.0','--port','8000'], cwd='/mnt/data/ikarus-ai-product-app/backend')
print('🚀 Uvicorn started. Use the public URL above.')
```

### Ingest dataset from Google Drive

```python
import requests
base = public_url
r = requests.post(base+'/ingest', json={'csv_path':'gdrive'})
print('Ingest response:', r.status_code, r.text)
```
