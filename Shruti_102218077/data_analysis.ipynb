# Data Analytics Notebook
Exploratory analysis of the furniture dataset with clear commentary.

```python
import pandas as pd, matplotlib.pyplot as plt
csv_path = '/content/data/products.csv'
df = pd.read_csv(csv_path)
print('Rows:', len(df)); df.head()
```

## Cleaning

```python
df['price'] = pd.to_numeric(df.get('price'), errors='coerce')
df['categories'] = df.get('categories').fillna('Unknown')
df['brand'] = df.get('brand').fillna('Unknown')
df['text_blob'] = (df.get('title','').fillna('') + ' | ' + df['brand'] + ' | ' + df.get('description','').fillna(''))
df.head(2)
```

## Price Distribution

```python
plt.figure()
df['price'].dropna().plot(kind='hist', bins=30)
plt.title('Price Distribution'); plt.xlabel('Price'); plt.ylabel('Count'); plt.show()
```

## Top Brands

```python
plt.figure(); df['brand'].value_counts().head(15).plot(kind='bar')
plt.title('Top Brands'); plt.ylabel('Count'); plt.show()
```

## Top Categories

```python
plt.figure(); cat = df['categories'].str.split(',').explode().str.strip()
cat.value_counts().head(15).plot(kind='bar')
plt.title('Top Categories'); plt.ylabel('Count'); plt.show()
```
